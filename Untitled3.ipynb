# Task
Analyze the breast cancer Wisconsin dataset by loading and preparing the data, performing feature normalization and correlation analysis, applying PCA, training and evaluating supervised models (SVM, Logistic Regression, and KNN) on both full and PCA-reduced features, comparing their performance, and visualizing decision boundaries. The data is located at `/kaggle/input/breast-cancer-wisconsin-data`.

## Load and Prepare Data

### Subtask:
Load the breast cancer Wisconsin dataset, identify features and the target variable, and perform any necessary initial data cleaning or encoding of categorical features.


**Reasoning**:
To identify the data file, I need to inspect the contents of the directory specified by the `path` variable. I will use `os.listdir()` for this purpose.



```python
import os

# List the contents of the directory to identify the data file
file_list = os.listdir(path)
print(f"Files in the dataset directory: {file_list}")
```

**Reasoning**:
Now that the data file has been identified as 'data.csv', I will load it into a pandas DataFrame and display the first few rows to understand its structure as per the instructions.



```python
import pandas as pd

# Construct the full path to the data file
data_file_path = os.path.join(path, 'data.csv')

# Load the dataset into a pandas DataFrame
df = pd.read_csv(data_file_path)

# Display the first few rows of the DataFrame
print("First 5 rows of the dataset:")
print(df.head())
```

**Reasoning**:
Based on the `df.head()` output and the subtask instructions, I will now identify the target variable (`diagnosis`), separate it from the features, drop the `id` and `Unnamed: 32` columns, and then encode the categorical target variable (`diagnosis`) into numerical values.



```python
import numpy as np

# Identify features (X) and target (y)
# The 'id' column is an identifier and 'Unnamed: 32' appears to be empty/unnecessary
X = df.drop(columns=['id', 'diagnosis', 'Unnamed: 32'])
y = df['diagnosis']

# Encode the categorical target variable 'diagnosis' (M=1, B=0)
y = y.map({'M': 1, 'B': 0})

print("First 5 rows of features (X):")
print(X.head())
print("\nShape of features (X):", X.shape)

print("\nFirst 5 rows of target (y):")
print(y.head())
print("\nShape of target (y):", y.shape)
```

**Reasoning**:
Now that features and target variables have been separated and encoded, the next step in data preparation is to check for any missing values in the features DataFrame (`X`) to ensure data quality and determine if any imputation or dropping strategies are needed.



```python
print("Missing values in features (X):")
print(X.isnull().sum())
```

## Feature Normalization and Correlation Analysis

### Subtask:
Apply a StandardScaler to normalize the numerical features of the dataset. Then, calculate and visualize the correlation matrix of these normalized features using a heatmap, ensuring to include a legend.


**Reasoning**:
To normalize the numerical features, calculate their correlation matrix, and visualize it as a heatmap, I will import necessary libraries, instantiate and apply `StandardScaler` to the features, then compute the correlation matrix, and finally use `seaborn` and `matplotlib` to generate and display the heatmap as per the instructions.



```python
from sklearn.preprocessing import StandardScaler
import seaborn as sns
import matplotlib.pyplot as plt

# Instantiate a StandardScaler object
scaler = StandardScaler()

# Apply the scaler to the feature DataFrame X
X_scaled = pd.DataFrame(scaler.fit_transform(X), columns=X.columns)

print("First 5 rows of scaled features (X_scaled):")
print(X_scaled.head())

# Calculate the correlation matrix of the X_scaled DataFrame
correlation_matrix = X_scaled.corr()

# Set up the matplotlib figure and axes for better visualization
plt.figure(figsize=(20, 18)) # Adjust figure size for better readability of many features

# Create a heatmap of the correlation matrix
sns.heatmap(correlation_matrix, cmap='coolwarm', annot=False, fmt=".2f", linewidths=.05, cbar=True)

# Add a title to the heatmap
plt.title("Correlation Matrix of Normalized Features", fontsize=20)

# Display the plot
plt.show()
```

## Split Data

### Subtask:
Split the preprocessed and normalized dataset into training and testing sets to prepare for model training and evaluation.


**Reasoning**:
To split the preprocessed and normalized dataset into training and testing sets, I will import the `train_test_split` function, apply it to `X_scaled` and `y` with the specified parameters, and then print the shapes of the resulting datasets to verify the split as per the instructions.



```python
from sklearn.model_selection import train_test_split

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.2, random_state=42, stratify=y
)

# Print the shapes of the resulting datasets
print("Shape of X_train:", X_train.shape)
print("Shape of X_test:", X_test.shape)
print("Shape of y_train:", y_train.shape)
print("Shape of y_test:", y_test.shape)
```

## Apply PCA

### Subtask:
Apply Principal Component Analysis (PCA) to the normalized training data. Generate both a 2-component PCA for decision boundary visualization and a higher-dimensional PCA (if deemed appropriate) for model training to capture sufficient variance.


**Reasoning**:
To apply PCA as instructed, I will import `PCA` from `sklearn.decomposition`, instantiate two PCA models (one with 2 components and another retaining 95% variance), fit them to the training data, transform both training and testing data, and then print the explained variance ratios and the shapes of the transformed datasets.



```python
from sklearn.decomposition import PCA

# 1. PCA for 2 components (for visualization)
pca_2_components = PCA(n_components=2)
X_train_pca2 = pca_2_components.fit_transform(X_train)
X_test_pca2 = pca_2_components.transform(X_test)

# 2. PCA for higher dimensions (retaining 95% variance)
pca_high_dim = PCA(n_components=0.95) # Retain 95% of variance
X_train_pca_high_dim = pca_high_dim.fit_transform(X_train)
X_test_pca_high_dim = pca_high_dim.transform(X_test)

# 3. Print explained variance ratio for both PCA models
print("Explained variance ratio for 2-component PCA:", pca_2_components.explained_variance_ratio_)
print("Cumulative explained variance for 2-component PCA:", sum(pca_2_components.explained_variance_ratio_))
print("\nExplained variance ratio for 95% variance PCA:", pca_high_dim.explained_variance_ratio_)
print("Cumulative explained variance for 95% variance PCA:", sum(pca_high_dim.explained_variance_ratio_))

# 4. Print the shapes of the newly created PCA-transformed datasets
print("\nShape of X_train_pca2:", X_train_pca2.shape)
print("Shape of X_test_pca2:", X_test_pca2.shape)
print("Shape of X_train_pca_high_dim:", X_train_pca_high_dim.shape)
print("Shape of X_test_pca_high_dim:", X_test_pca_high_dim.shape)
```

## Train and Evaluate Supervised Models (Full Features)

### Subtask:
Train Support Vector Machine (SVM), Logistic Regression, and K-Nearest Neighbors (KNN) models using the full set of normalized features. Evaluate each model's performance on the test set and report metrics such as accuracy, precision, recall, and F1-score.


**Reasoning**:
To begin training and evaluating the models, I will import the necessary model classes and evaluation metrics from scikit-learn as per the instructions, and initialize a dictionary to store the performance results.



```python
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

# Initialize a dictionary to store the performance metrics for models trained on full features
performance_metrics_full_features = {}
print("Initialized performance_metrics_full_features dictionary.")
```

**Reasoning**:
I will now instantiate, train, predict, and evaluate each of the specified models (SVM, Logistic Regression, KNN) using the full set of normalized features. I will store their performance metrics (accuracy, precision, recall, F1-score) in the `performance_metrics_full_features` dictionary and then print the dictionary.



```python
print("Training and evaluating models on full features...")

# 1. Support Vector Machine (SVM)
svm_model = SVC(random_state=42)
svm_model.fit(X_train, y_train)
y_pred_svm = svm_model.predict(X_test)

performance_metrics_full_features['SVM'] = {
    'Accuracy': accuracy_score(y_test, y_pred_svm),
    'Precision': precision_score(y_test, y_pred_svm),
    'Recall': recall_score(y_test, y_pred_svm),
    'F1-Score': f1_score(y_test, y_pred_svm)
}

# 2. Logistic Regression
log_reg_model = LogisticRegression(random_state=42, solver='liblinear') # 'liblinear' solver for small datasets
log_reg_model.fit(X_train, y_train)
y_pred_log_reg = log_reg_model.predict(X_test)

performance_metrics_full_features['Logistic Regression'] = {
    'Accuracy': accuracy_score(y_test, y_pred_log_reg),
    'Precision': precision_score(y_test, y_pred_log_reg),
    'Recall': recall_score(y_test, y_pred_log_reg),
    'F1-Score': f1_score(y_test, y_pred_log_reg)
}

# 3. K-Nearest Neighbors (KNN)
knn_model = KNeighborsClassifier()
knn_model.fit(X_train, y_train)
y_pred_knn = knn_model.predict(X_test)

performance_metrics_full_features['KNN'] = {
    'Accuracy': accuracy_score(y_test, y_pred_knn),
    'Precision': precision_score(y_test, y_pred_knn),
    'Recall': recall_score(y_test, y_pred_knn),
    'F1-Score': f1_score(y_test, y_pred_knn)
}

# Print the results
print("\nPerformance Metrics (Full Features):")
for model_name, metrics in performance_metrics_full_features.items():
    print(f"\n{model_name}:")
    for metric_name, value in metrics.items():
        print(f"  {metric_name}: {value:.4f}")
```

## Train and Evaluate Supervised Models (PCA-Reduced Features)

### Subtask:
Train Support Vector Machine (SVM), Logistic Regression, and K-Nearest Neighbors (KNN) models using the PCA-reduced features (from the higher-dimensional PCA). Evaluate each model's performance on the test set and report the same metrics.


**Reasoning**:
To begin training and evaluating the models with PCA-reduced features, I will first initialize a new dictionary to store their performance metrics, as specified in the instructions. This will keep the results separate from the models trained on full features.



```python
performance_metrics_pca_features = {}
print("Initialized performance_metrics_pca_features dictionary.")
```

**Reasoning**:
I will now instantiate, train, predict, and evaluate each of the specified models (SVM, Logistic Regression, KNN) using the higher-dimensional PCA-reduced features. I will store their performance metrics (accuracy, precision, recall, F1-score) in the `performance_metrics_pca_features` dictionary and then print the dictionary.



```python
print("Training and evaluating models on PCA-reduced features (higher-dimensional)...")

# 1. Support Vector Machine (SVM) with PCA-reduced features
svm_pca_model = SVC(random_state=42)
svm_pca_model.fit(X_train_pca_high_dim, y_train)
y_pred_svm_pca = svm_pca_model.predict(X_test_pca_high_dim)

performance_metrics_pca_features['SVM_PCA'] = {
    'Accuracy': accuracy_score(y_test, y_pred_svm_pca),
    'Precision': precision_score(y_test, y_pred_svm_pca),
    'Recall': recall_score(y_test, y_pred_svm_pca),
    'F1-Score': f1_score(y_test, y_pred_svm_pca)
}

# 2. Logistic Regression with PCA-reduced features
log_reg_pca_model = LogisticRegression(random_state=42, solver='liblinear')
log_reg_pca_model.fit(X_train_pca_high_dim, y_train)
y_pred_log_reg_pca = log_reg_pca_model.predict(X_test_pca_high_dim)

performance_metrics_pca_features['Logistic Regression_PCA'] = {
    'Accuracy': accuracy_score(y_test, y_pred_log_reg_pca),
    'Precision': precision_score(y_test, y_pred_log_reg_pca),
    'Recall': recall_score(y_test, y_pred_log_reg_pca),
    'F1-Score': f1_score(y_test, y_pred_log_reg_pca)
}

# 3. K-Nearest Neighbors (KNN) with PCA-reduced features
knn_pca_model = KNeighborsClassifier()
knn_pca_model.fit(X_train_pca_high_dim, y_train)
y_pred_knn_pca = knn_pca_model.predict(X_test_pca_high_dim)

performance_metrics_pca_features['KNN_PCA'] = {
    'Accuracy': accuracy_score(y_test, y_pred_knn_pca),
    'Precision': precision_score(y_test, y_pred_knn_pca),
    'Recall': recall_score(y_test, y_pred_knn_pca),
    'F1-Score': f1_score(y_test, y_pred_knn_pca)
}

# Print the results
print("\nPerformance Metrics (PCA-Reduced Features - Higher-Dimensional):")
for model_name, metrics in performance_metrics_pca_features.items():
    print(f"\n{model_name}:")
    for metric_name, value in metrics.items():
        print(f"  {metric_name}: {value:.4f}")
```

## Compare Model Performance

### Subtask:
Create a visualization, such as a bar chart, to compare the performance metrics (e.g., accuracy) of all trained models across both full features and PCA-reduced features. Include a legend to distinguish between models and feature sets.


**Reasoning**:
To compare the performance metrics as instructed, I will first combine the existing performance dictionaries into a single pandas DataFrame, making it suitable for plotting. Then, I will use `seaborn` and `matplotlib.pyplot` to create a bar chart specifically for 'Accuracy', ensuring proper labeling, titles, and a legend for clarity.



```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Initialize an empty list to store all performance data
all_performance_data = []

# Process performance metrics for full features
for model_name, metrics in performance_metrics_full_features.items():
    data = {'Model': model_name, 'Feature Set': 'Full Features'}
    data.update(metrics)
    all_performance_data.append(data)

# Process performance metrics for PCA-reduced features
for model_name, metrics in performance_metrics_pca_features.items():
    # Remove '_PCA' suffix from model name for consistent labeling
    clean_model_name = model_name.replace('_PCA', '')
    data = {'Model': clean_model_name, 'Feature Set': 'PCA-Reduced Features'}
    data.update(metrics)
    all_performance_data.append(data)

# Create a DataFrame from the combined data
performance_df = pd.DataFrame(all_performance_data)

print("Combined Performance Metrics DataFrame:")
print(performance_df)

# Create a bar chart to compare Accuracy
plt.figure(figsize=(10, 6))
sns.barplot(x='Model', y='Accuracy', hue='Feature Set', data=performance_df, palette='viridis')

plt.title('Model Performance Comparison (Accuracy)', fontsize=16)
plt.xlabel('Model Type', fontsize=12)
plt.ylabel('Accuracy', fontsize=12)
plt.ylim(0.9, 1.0) # Set y-axis limit for better comparison of high accuracies
plt.legend(title='Feature Set')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
